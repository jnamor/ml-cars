```python
import requests
import pandas as pd
```

```python
def build_url(values):
    return "/".join(['%s' % values[key] for key in values])
```

```python
values = {
    'location': 75000,
    'version': 'volvo-xc90',
    'year': 2019,
    'mileage': 10000,
    'fuel_type': 'essence',
    'emission': 155,
    'transmission': 'manuelle',
    'door_nb': 4,
    'technical_power': 7,
    'seat_nb': 5,
    'actual_power': 130,
    'body_colour': 'noir',
    'body_type': '4x4'
}
```

```python
url = f"http://127.0.0.1:5000/prediction/{build_url(values)}"
response = requests.get(url)
print(f"Le prix prédit est égal à {response.text.strip()}€")
```

## Optimization

```python
ref_float = pd.read_csv('data/agg_paruvendu.csv')
ref_float.head()
```

```python
ref_object = pd.read_csv("data/paruvendu.csv")
ref_object['version'] = ref_object['version'].str.replace('/', '-')
ref_object.head()
```

```python
values = {
    'location': 75000,
    'version': 'volvo-xc90',
    'year': 2019,
    'mileage': 10000,
    'fuel_type': 'essence',
    'emission': 155,
    'transmission': 'manuelle',
    
    # Facultatif
    'door_nb': '',
    'technical_power': '',
    'seat_nb': '',
    'actual_power': '',
    'body_colour': '',
    'body_type': ''
}

for key, value in values.items():
    if value == '':
        if key in ['door_nb', 'technical_power', 'seat_nb', 'actual_power']:
            values[key] = int(ref_float[ref_float['version'] == values['version']][key].item())
        else:
            base = ref_object[ref_object['version'] == values['version']][key]
            values[key] = base.value_counts().index[0]
```

```python
values
```

```python
url = f"http://127.0.0.1:5000/prediction/{build_url(values)}"
response = requests.get(url)
print(f"Le prix prédit est égal à {response.text.strip()}€")
```
